```python
import requests
from bs4 import BeautifulSoup

page = requests.get("https://gcoea.ac.in/")
soup = BeautifulSoup(page.content,'html.parser')
#print(soup)
title = soup.title
#print(title)

```

```python
dir_class = soup.find_all(class_="directorblock")
print(dir_class)
```

```python
anchorTags = soup.find_all('a')
#print(anchorTags)
```

```python
container_class=(soup.find_all(class_='container'))
print(container_class)
```

```python
all_text = soup.get_text()
print(all_text)
```

```python
for links in anchorTags:
    print(links.get('href'))
```

```python
list1 = [soup.find(class_="directorblock").get_text() for i in dir_class]
print(list1)
```

```python
list2 = [soup.find(class_='container').get_text() for i in container_class[::] ]
print(list2)
```
